```python
# code A
def max_sum(n, k, arr):
    if k == n:
        return sum(arr)
    else:
        _sum = sum(arr[0:k])
        max_sum = _sum
        for i in range(k):
            _sum = _sum - arr[k - i - 1] + arr[n - i - 1]
            if _sum > max_sum:
                max_sum = _sum
        return max_sum
            
if __name__ == '__main__':
    n = int(input())
    k = int(input())
    arr = list(map(int, input().split()))
    
    print(max_sum(n, k, arr))
```
