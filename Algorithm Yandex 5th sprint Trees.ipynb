## Деревья. Вступление.
[телепорт в оглавление](#title)

В прошлом спринте мы говорили о хеш-таблицах, которые помогают искать, добавлять и удалять данные за O(h)+O(1+α)O(h) + O(1 + \alpha)O(h)+O(1+α), где hhh — сложность вычисления хеш-функции, а α\alphaα — коэффициент заполнения хеш-таблицы. Платить за такую привлекательную сложность приходится дополнительной памятью для хранения пустых корзин.
Но есть задачи, для решения которых необходимо получить высокую скорость поиска информации. Например, когда нужно быстро находить объекты в огромной базе данных. Записей в базе будет так много, что для поддержания α\alphaα в рекомендуемых границах придётся потратить слишком много дополнительной памяти. Поэтому при решении подобных задач разработчики используют не хеш-таблицу, а совсем другую, более подходящую для этого структуру данных — она называется «дерево».
В этом спринте мы расскажем, как устроены деревья и как с их помощью можно эффективно решить задачу поиска элемента по ключу. Кроме того, мы поговорим о том, как деревья помогают находить все объекты, ключи которых попадают в заданный интервал. Например, вы узнаете, как быстро получить список всех фильмов продолжительностью больше 70 минут, но меньше 90. А ещё мы расскажем, как на основе деревьев реализовать очередь с элементами разной важности и как при помощи такой очереди отсортировать массив.
Дерево как иерархическая структура
image
Тимофей
Деревья — это не только знакомые всем нам большие растения, которыми мы любуемся в садах, лесах и парках. Это ещё и интересный способ представлять информацию в виде иерархической структуры, где элементы, находящиеся ниже по иерархии, зависят от элементов, расположенных выше.
Например, у организации, в которой я работаю, есть определённая структура: дерево сотрудников. Выше всех находится генеральный директор, под ним — руководители подразделений, ниже — руководители команд, ещё ниже — рядовые сотрудники. У каждого в этой структуре есть ровно один непосредственный руководитель, а вот подчинённых может быть сразу несколько или вообще ни одного.
Рита
В школе на уроках биологии нам рассказывали про общую систему классификации животных, растений и всех живых существ. Она построена по принципу иерархии: насколько я помню, там сначала идёт царство, потом от него ответвляется несколько отделов, а отделы в свою очередь дробятся на классы и так далее. Это ведь тоже дерево?
Тимофей
Да, это прекрасный пример дерева!
image
Элементы на верхнем уровне дерева обобщённо описывают элементы, к которым у них есть стрелки на нижнем уровне. И высшие споровые растения, и семенные растения — это частные случаи растений из отдела высших растений
Если схема достаточно полная, то мы можем узнать список всех растений определённого семейства. И наоборот, найдя какое-то интересующее нас растение, сможем сказать, к какому роду оно принадлежит. Это очень удобное свойство иерархических структур!
Основные термины
Прежде чем говорить о том, как создавать деревья и работать с ними, давайте договоримся о терминах. Их довольно много, и у большинства есть синонимы (мы перечислим и их тоже). В случае, если при прохождении следующих уроков вы обнаружите, что не помните значение того или иного термина — смело возвращайтесь к этой странице!
Давайте посмотрим, из чего состоят деревья, на примере иерархии сотрудников, о которой говорил Тимофей.

    Вершина дерева — объект, который может хранить информацию и ссылаться на другие вершины. В примере Тимофея вершинами являются все сотрудники. Синонимы: узел, нода.
    Ребро дерева — направленная, то есть иерархическая, связь между двумя вершинами. Посмотрев на направление рёбер дерева, можно однозначно определить, какая из двух связанных вершин выше по иерархии, а какая — ниже.
    Путь в дереве — набор вершин, в котором следующие друг за другом узлы соединены рёбрами. Так как рёбра имеют строго одно направление, то и путь возможно проложить только в одну сторону — сверху вниз по иерархии, и никак иначе.
    Синоним: маршрут.
    Дерево — совокупность вершин и рёбер без циклов. Если из вершины X1X_1X1​ существует путь до вершины X2X_2X2​, то гарантируется, что не существует пути от X2X_2X2​ к X1X_1X1​. В примере Тимофея сотрудник не может быть одновременно и подчинённым, и руководителем для одного и того же человека.

Подробнее о вершинах

    У каждой вершины есть не более одного родителя. Если у вершины нет родителя, она называется корнем. В нашем примере родителем является любой руководитель любого сотрудника. А корень — это генеральный директор.

image
Эта структура не является деревом, так как у вершины 7 оказалось сразу два родителя

    Все вершины, через которые сверху вниз проходит путь от корня до вершины XXX, называются предками вершины XXX. То есть это родитель, родитель родителя, родитель родителя родителя и так далее. Непосредственный предок — это родитель. Генеральный директор в терминологии деревьев является предком для всех сотрудников.
    У вершин могут быть дети в количестве от 0 до бесконечности. В примере Тимофея дети — это подчинённые сотрудника. Синонимы: ребёнок, дочерние узлы, непосредственные потомки.

image
Вершина дерева «Руководитель отдела разработки Т.В. Ёлкин» ссылается на своего единственного потомка — вершину «Руководитель отдела стажёров Тимофей». Между этими вершинами есть направленное ребро, обозначающее, что Тимофей подчиняется Т.В. Ёлкину. В свою очередь, у Т.В. Ёлкина один родитель — «Генеральный директор А.П. Дубов». Это общий предок и для Т.В. Ёлкина, и для Тимофея.

    Все дети одного родителя называются по отношению друг к другу sibling.
    Все вершины, до которых есть путь от вершины YYY, называются потомками вершины YYY.
    Если у вершины нет дочерних узлов, она называется листом. В примере Тимофея листья — это рядовые сотрудники, у которых нет подчинённых. Синоним: терминальная вершина.
    Высота дерева с корнем — это количество вершин в самом длинном пути от корня до листа. Синоним: глубина.

Виды деревьев
В общем случае количество дочерних узлов в дереве может быть произвольным. А если в рамках какой-нибудь задачи мы ограничили максимальное число потомков, то это число называется «арностью дерева».
image
Несмотря на то, что у некоторых узлов есть только по одному ребёнку, это дерево арности 3, так как 3 — это максимальное количество детей у вершин этого дерева
Если каждый узел в дереве имеет не более двух потомков, такое дерево называют бинарным, или двоичным.
image
Бинарное дерево. У каждой вершины — не более двух поддеревьев
На этом курсе вы уже встречались со структурой данных, которая является деревом. Как думаете, что это было?
Массив.
Хеш-таблица.
Правильный ответ
Односвязный список.
Односвязный список — дерево, в котором у всех вершин есть ровно один ребёнок. Вершины — это элементы списка, рёбра — ссылка next от одного элемента к другому. А голова списка является корнем дерева.
image
Ни одна из этих структур данных не подходит под определение дерева.
Представление деревьев в программе
Как и элементы в связных списках, узлы деревьев представляют в виде класса или структуры — в зависимости от языка программирования.
Отличие в том, что в дереве у вершины может быть сколько угодно детей, для их хранения удобно завести в классе массив children:

class Node:
    def __init__(self, obj, children=None):
        self.obj = obj
        if children is not None:
            self.children = children
        else:
            self.children = [] 

Разберём частный случай деревьев — бинарное дерево. В случае с бинарным деревом, как мы уже говорили, детей может быть от 0 до 2. Обычно их не объединяют в массив, а объявляют как два отдельных поля:

class Node:
    def __init__(self, obj, left=None, right=None):
        self.obj = obj
        self.left = left
        self.right = right 

Именно бинарные деревья наиболее популярны в программировании. В этом спринте мы будем говорить только о бинарных деревьях.

## Двоичные деревья поиска
[телепорт в оглавление](#title)

Представьте, что у вас регулярно возникает такая задача: нужно находить должность сотрудника по его фамилии. Чтобы найти необходимую запись в произвольном дереве, надо последовательно обойти все его вершины. Так происходит из-за того, что у нас нет никаких подсказок, где искать нужного сотрудника. Он может оказаться в какой угодно части дерева. Поэтому поиск по произвольному дереву в худшем случае будет работать за O(n)O(n)O(n). Столько же времени занял бы перебор всех элементов в массиве.
Бинарное дерево поиска
Как мы выяснили, произвольные деревья плохо справляются с задачей поиска: для этого они слишком медленные. Но стоит добавить несколько дополнительных условий, и обычное дерево превратится в дерево поиска(англ. binary search tree, BST), а скорость операций резко возрастёт. Вот эти условия:

    Если в узле дерева записано значение ХХХ, то в левом поддереве располагаются только узлы со значениями, меньшими или равными ХХХ.
    Если в узле дерева записано значение ХХХ, то в правом поддереве располагаются только узлы со значениями, большими или равными ХХХ.
    Левое и правое поддеревья отвечают тем же двум условиям. Это рекурсивное правило. Из него следует, что любое поддерево в дереве поиска будет являться деревом поиска.

image
Отличить дерево поиска от обычного — непросто. Для этого надо сравнить информацию, записанную в корне, с данными во всех его детях.
image
Только у одного из этих деревьев числа в узлах расположены в том порядке, о котором мы говорили выше
Деревья поиска и ключи
Быстро узнавать, содержится ли в базе данных элемент с определёнными свойствами — это отличное умение, но одного его недостаточно для решения прикладных задач. Например, таких:

    В большой базе данных в виде дерева поиска представлена информация о самых разных фильмах. Необходимо по названию фильма узнать его жанр и год выпуска.
    Данные о пользователях соцсети хранятся в дереве поиска. Необходимо по числовому ID владельца страницы узнать его возраст и дату регистрации в этой соцсети.

Дерево поиска в таких ситуациях строится по ключу — то есть определённому набору свойств элементов. Для ключей предусмотрен компаратор: в случае с числами это простой оператор сравнения «меньше» (<). Мы уже сталкивались с компараторами, когда говорили о сортировках. В задачах, описанных выше, ключами будут название фильма (строка) и ID пользователя (число).
Обратите внимание, что деревья поиска позволяют хранить и повторяющиеся ключи. Например, в базе данных может находиться сразу несколько фильмов с названием «Игра». При этом повторяющиеся ключи не приведут к коллизии, так как расположение вершины в памяти не зафиксировано, а значит, каждый из повторяющихся элементов найдёт своё место.
Во всех дальнейших иллюстрациях числа в вершинах деревьев — это ключи. По ключу мы можем находить элементы (например, конкретный фильм — по его названию) и узнавать их поля, если они есть (например, год создания этого фильма или жанр).
Всегда ли можно однозначно задать дерево поиска по неупорядоченному набору ключей?
Да, дерево поиска строится однозначно.
Тоже правильный ответ
Нет, одному набору ключей соответствует много разных деревьев поиска.
Почти всегда дерево поиска строится однозначно, но если в наборе есть одинаковые ключи, то получится больше одного дерева поиска.
Допустим, у нас есть набор из трёх ключей: 1, 2, 3. В зависимости от того, в каком порядке мы будем добавлять эти ключи, у нас будут получаться разные деревья поиска.
image
Для набора из 3 элементов есть 6 вариантов последовательностей добавления ключей в дерево поиска. Но в вариантах 3 и 4 деревья поиска получатся одинаковые
В бинарном дереве поиска высоты HHH расположено nnn вершин. Какая сложность поиска в таком дереве в худшем случае?
O(1)O(1)O(1)
Правильный ответ
O(H)O(H)O(H)
O(log⁡n)O(\log{n})O(logn)
Тоже правильный ответ
O(n)O(n)O(n)
Да. Если у каждой вершины бинарного дерева ровно один ребёнок, то его высота и число элементов в нём совпадают.
O(n⋅log⁡n)O(n\cdot\log{n})O(n⋅logn)
O(n⋅H)O(n \cdot H)O(n⋅H)
В следующих уроках расскажем о том, как поддерживать небольшую высоту HHH, чтобы поиск по дереву работал быстро. А пока будем измерять скорость операций по высоте дерева, а не по количеству вершин в нём.
Поиск по дереву
Рассмотрим, как работает поиск в случае, если нам надо найти какой-нибудь элемент с заданным ключом.
Благодаря особой структуре дерева поиска работать с ним удобнее, чем с обычным деревом. Каждый раз, когда мы выбираем, в каком поддереве продолжать поиск, у нас есть указатель-подсказка: если элемент, записанный в текущей вершине, больше искомого, то нужно продолжить поиск в левом поддереве, иначе — в правом. А если ключ вершины равен искомому, то поиск успешно завершается.
image
Пусть мы ищем вершину 34. Нам потребуется провести 4 сравнения:

    34 == 3? Нет. 34 < 3? Нет, значит, идём в правое поддерево.
    34 == 12? Нет 34 < 12? Нет, значит, идём в правое поддерево.
    34 == 99? Нет 34 < 99? Да, значит, идём в левое поддерево.
    34 == 34? Да, мы нашли нужную вершину!

Но как понять, что искомой вершины в дереве поиска нет? Если мы хотим продолжить поиск в одном из поддеревьев, а его нет, и вместо ссылки на него записан None, значит, искомый элемент отсутствует.
image
Пусть мы ищем вершину 5. Нам потребуется провести 3 сравнения:

    5 == 3? Нет. 5 < 3? Нет, значит, идём в правое поддерево.
    5 == 12? Нет 5 < 12? Да, значит, идём в левое поддерево.
    5 == 7? Нет. 5 < 7? Да, значит, идём в левое поддерево. Ой, его нет! Значит, вершина 5 в дереве отсутствует.

Запишем это псевдокодом:

    функция find_node(root, value):
      Если мы пришли в поддерево, а его не существует, значит, нужного элемента в дереве поиска нет
      если root == None, то верни "Значение {value} не найдено"
      если value < root.value, то верни find_node(root.left, value)
      если value = root.value, то верни root 
      если value > root.value, то верни find_node(root.right, value)
      верни результат 

Если глубина рекурсии слишком большая, стек вызовов переполнится. Как этого избежать?
В нашем случае глубина рекурсии зависит от высоты дерева. Если дерево слишком высокое, может случиться проблема. Попробуй переписать рекурсивную функцию поиска через цикл while, чтобы избежать переполнения стека.

## Обход дерева
[телепорт в оглавление](#title)

Как мы уже говорили, деревья удобны для хранения данных с иерархической структурой. Но когда мы храним какой-то элемент, нам важно при необходимости вывести этот элемент на экран и отобразить его место в иерархической структуре. В случае с деревом организации мы можем, например:

    вывести список рабочих номеров телефонов всех сотрудников;
    вывести несколько списков, сгруппировав сотрудников по подразделениям или отделам, в которых они работают.

Но так как, в отличие от массивов и связных списков, из вершины дерева мы можем попасть сразу в несколько других вершин, важно определить порядок обхода элементов. Это необходимо, чтобы:

    посетить каждый элемент ровно один раз,
    не оставить ни один элемент без внимания.

Для решения этой задачи разработчики используют стандартные методы обхода деревьев.
Прямой и обратный обходы
Эти два обхода применимы для всех видов деревьев — вне зависимости от их арности и дополнительных ограничений (например, на расположение данных).
Прямой обход описывает порядок, которому интуитивно следует большинство людей: сначала мы смотрим на корень дерева, потом переводим взгляд на поддерево. В поддеревьях рекурсивно повторяем тот же алгоритм: сначала изучаем корень, потом — поддеревья.
Порядок обхода детей не регламентируется, но так как для европейцев привычно читать текст слева направо, мы рассматриваем поддеревья аналогично — слева направо.
Если у нас есть структура организации, сохранённая в виде дерева, то прямой обход — самый естественный способ вывести её на экран:
image
Псевдокод для прямого обхода выглядит так:

функция print_forward(vertex):
  print(vertex.value)
  для каждого ребёнка child из vertex.children:
    print_forward(child) 

Второй стандартный метод обхода дерева — обратный обход. Он очень похож на прямой. Единственное различие между ними в том, что при обратном обходе сначала рассматриваются все поддеревья в любом порядке и только потом — корень поддерева.
С помощью метода обратного обхода можно вывести на экран список задач по проекту в порядке их выполнения. В корне записана главная цель, в детях — подзадачи. Некоторые подзадачи тоже будут составными проектами. При этом метод обратного обхода гарантирует, что можно выполнять текущую задачу, если все задачи, написанные выше, выполнены:
image
Псевдокод реализации обратного порядка выглядит так:

функция print_reversed(vertex)
  для каждого ребёнка child из vertex.children
    print_reversed(child)
  print(vertex.value) 

При прямом обходе мы обрабатываем узел, когда приходим в него впервые, а при обратном — когда уходим из него навсегда.
Центрированный обход
Центрированный метод обхода подходит только для бинарных деревьев. Поскольку подразумевает, что мы знаем, где находится центральная ось каждого из поддеревьев и всего дерева в целом. Суть метода в том, что мы сначала выводим все элементы одного поддерева. Потом корень. Потом все элементы второго поддерева, если они есть. При этом для каждого поддерева мы производим вывод вершин таким же алгоритмом. То есть рекурсивно.
В этом методе есть два варианта:

    сначала выводить левое поддерево, потом центр, потом правое: такой обход по сокращению букв называют LMR (left-middle-right);
    сначала правое поддерево, потом центр, потом левое — RML(right-middle-left).

image
Центрированный обход дерева: сначала обойдём все элементы левого поддерева а порядке ABCDE, потом корневую вершину F, потом все вершины правого поддерева в порядке GHI
Псевдокод для центрированного подхода выглядит так:

функция print_LMR(vertex)
  если vertex.left != None тогда print_LMR(vertex.left)
  print(vertex.value)
  если vertex.right != None тогда print_LMR(vertex.right) 

Обходы дерева поиска
Обойти и вывести на экран можно любое дерево. Но в случае, когда мы производим обход бинарного дерева поиска, мы получаем полезный бонус.
Так как для каждого поддерева в дереве поиска все вершины в левом поддереве меньше корня и меньше вершин в правом поддереве, центрированный обход дерева выведет все элементы дерева отсортированными:

    при LMR-обходе данные будут отсортированы от меньшего элемента к большему,
    при RML-обходе данные будут отсортированы от большего элемента к меньшему.

Онлайн-задача поиска
В программировании принято делить алгоритмы на работающие офлайн и онлайн.
Первые получают на вход сразу все данные, один раз их обрабатывают и выдают ответ на какой-то вопрос. Вторые получают данные постепенно, а ответ на один и тот же вопрос запрашивается у них периодически. При этом программа для получения ответа должна опираться на всю информацию, известную ей к этому моменту. Ожидается, что, получив новые данные, программа эффективно дообучается и модифицирует старый ответ для вычисления нового.
Мы уже рассматривали алгоритм бинарного поиска, который в худшем случае работает за O(log⁡n)O(\log{n})O(logn). Напоминаем, что данные в массиве должны быть отсортированы. Алгоритм двоичного поиска хорошо работает в офлайн-режиме, то есть когда все данные известны заранее. Если мы добавляем в массив новые элементы, нам требуется снова пересортировывать весь массив.
Бинарное дерево поиска решает ту же задачу, оно позволяет найти элемент среди множества других за O(H)O(H)O(H), но, в отличие от бинарного поиска, данные в бинарном дереве поиска отсортированы в любой момент времени. И найти ключ в таком дереве можно, не производя дополнительных подготовительных операций. Плата за возможность поиска ключей в онлайн режиме –– более высокая сложность добавления элемента.
Поиск элементов в интервале
Бинарное дерево поиска можно использовать для того, чтобы искать все элементы, у которых ключ располагается в некотором интервале. Допустим, у нас есть дерево поиска товаров в магазине, упорядоченное по их стоимости. Давайте разберёмся, как найти все товары с ценой от 1500 до 1900 рублей.
Например, в нашем дереве будут такие узлы: футбольный мяч — 900 ₽, бумеранг — 1400 ₽, воздушный змей — 1550 ₽, ракетки для игры в бадминтон — 1600 ₽, ракетки для игры в настольный теннис — 1600 ₽, гантели — 1800 ₽, коврик для йоги — 2000 ₽, кроссовки — 2200 ₽.
image
Первым делом нужно среди товаров стоимостью больше 1500 рублей найти самый дешёвый. В нашем случае это воздушный змей — он стоит 1550 рублей. Если бы были ещё товары по 1550, нам следовало бы найти тот из них, который в дереве расположен левее других (так как в порядке обхода самый левый элемент будет посещён раньше всех остальных). Для этого мы применим стандартный алгоритм поиска с небольшими модификациями.

    Если корень дерева меньше искомого элемента 1500, поиск производится в правом поддереве.
    Если корень дерева больше или равен искомому элементу 1500, мы рекурсивно повторяем процедуру поиска в левой половине.
    Если в левой половине элементов не нашлось, значит, корень дерева является первым элементом, который имеет стоимость больше либо равную 1500.

Может так получиться, что в процессе работы алгоритма мы спустимся до узла, который меньше искомого. Например, в корне дерева значение оказалось большее искомого, а в левом поддереве все элементы, напротив, меньше искомого.
В нашем примере алгоритм спустится до узла, который меньше искомого — до товара со стоимостью 1400. Однако следующий же элемент в порядке LMR-обхода будет иметь стоимость не меньше 1500. Чтобы получить этот элемент, нам потребуется найти в дереве узел, следующий по порядку за данным.
После того как мы нашли первый элемент, который будет правее левой границы — 1550, нам следует обходить все элементы дерева в порядке возрастания, пока они не превзойдут правую границу. Итого в нашем интервале оказалось четыре элемента: 1550, 1600, 1600 и 1800.
Поиск первого элемента занимает O(H)O(H)O(H) операций. Перебор последующих значений линейно зависит от количества элементов в интервале.

## Двоичные деревья поиска. Вставка
[телепорт обратно в оглавление](#title)

- Если нужно записать новый или удалить старый элемент?
- Эти операции похожи на поиск! Элемент нужно добавить туда, где ты его будешь искать, когда он понадобится. Так же, как при поиске, мы спускаемся по дереву до подходящего места и добавляем новый узел.
- А как мы поймём, что нашли нужное место?
- Этим местом будет незаполненная позиция. Давай разберём на примере. Попробуем добавить элемент с ключом 39 в такое дерево:
image

    Начинаем в корне. Там записан ключ 35. Сравниваем 39 с 35 и понимаем, что узел следует добавлять в правое поддерево, ведь именно там мы в дальнейшем будем искать элемент с ключом 39.
    Теперь сравниваем 39 и 42. На этот раз правило поиска говорит, что узел следует искать в левом поддереве, корнем которого является узел 37.
    Добавляемый ключ 39 больше, чем 37, значит, узел должен быть добавлен в правое поддерево. Но правого поддерева пока что не существует, в качестве правой ветви указан None. Значит, на этом месте мы и создадим узел, то есть создадим экземпляр класса Node c ключом 39, и запишем его правым ребёнком вершины 37.

Гоша
А что если мы попытаемся добавить ключ, который уже есть в дереве?
Тимофей
Если ключ есть, то нам надо найти подходящее место для второго такого же ключа.
По определению, если значение ключа равно значению в корне, то мы его можем расположить как в левом, так и в правом поддеревьях. Но для простоты реализации будем считать, что в левое поддерево мы кладём элементы меньше корня, а в правое — большие или равные корню.
Будем спускаться по дереву до тех пор, пока не дойдём до места, в котором отсутствует поддерево, и именно в это место добавим новую вершину с повторяющимся ключом. Давай рассмотрим этот алгоритм на примере:
image
Допустим, нам надо вставить ключ 6 в данное дерево. Наши действия:

    Начинаем в корне. Там записан ключ 6. Добавляемый ключ больше или равен значению в корне, значит, идём в правое поддерево.
    В поддереве корень 7. 7 > 6, продолжаем поиск места под новый ключ в левом поддереве.
    Корень левого поддерева опять 6, значит, мы опять должны пойти в правое поддерево, но его нет: вместо него сейчас записан None. Значит, именно в этом месте надо создать новую вершину.

Гоша
А всегда ли найдётся подходящее место?
Тимофей
Да, найдётся. На каждом шаге мы всякий раз принимаем решение: спускаться на один уровень ниже в правое поддерево или в левое, то есть не стоим на месте. А поскольку дерево имеет конечную глубину, то рано или поздно алгоритм закончит свою работу. А это возможно только на пустой вершине.
Гоша
Понял. Значит, мы спускаемся по дереву, пока не встретим пустое место. И в это место добавляем элемент. А как это лучше всего реализовать?
Тимофей
Можно сделать это с помощью рекурсии. Вставка элемента в корень дерева описывается как вставка либо в левое, либо в правое поддерево. На каждом следующем шаге рекурсии дерево будет уменьшаться. Если поддерево отсутствует, мы создадим и прикрепим к дереву узел — на этом рекурсия завершится. Осталось только написать псевдокод:

функция insert_node(root, key):
  если key < root.key, то 
      если root.left == None, то root.left = Node(key)
    иначе insert_node(root.left, key)
  если key >= root.key, то 
      если root.right == None, то root.right = Node(key)
    insert_node(root.right, key)  

Гоша
Здорово! Это действительно похоже на поиск элемента в дереве. Осталось разобраться с удалением, и я стану специалистом по деревьям!
Тимофей
С удалением всё немного сложнее. Я расскажу об этом в следующий раз.

## Двоичные деревья поиска. Удаление
[телепорт в оглавление](#title)

Когда удаляется вершина, каждое ребро, которое в неё входило или из неё исходило, становится невалидным (недопустимым). Оно как бы повисает в воздухе, одним из своих концов указывая на отсутствующий объект. Поэтому такие рёбра удаляются вместе с вершиной.
Если из двоичного дерева удалить вершину и все связанные с ней рёбра, сколько отдельных деревьев может получиться?

0
Правильный ответ

Правильный ответ

Правильный ответ

Тимофей
У вершины в бинарном дереве может быть до трёх рёбер (одно — от родителя к вершине и два — от вершины к детям). Соответственно, она может объединять собой три отдельные группы вершин. Удаление такой вершины разорвёт связь между этими группами.
Гоша
Ого, сколько вариантов! Давай это нарисуем, я не уверен, что представляю, как каждый из них может случиться.
image

    Если дерево состояло из одной вершины, то после её удаления дерева не останется.
    Если мы удаляем лист, то дерево останется одним деревом и не распадётся на части.
    Если мы удаляем корень, у которого есть оба поддерева, то каждое поддерево станет отдельным деревом.
    Если мы удаляем вершину, у которой есть оба ребёнка и родитель, то дерево распадётся на родительское и два поддерева.

Тимофей
При наивном удалении вершины дерево распадается. Чтобы этого избежать, надо соединить распавшиеся поддеревья новыми рёбрами. Давай посмотрим, как это можно сделать.
Объединение распадающихся поддеревьев
Для того чтобы дерево осталось деревом поиска, на место удалённой вершины надо поставить какую-то другую вершину.
Для начала из всех отсоединяющихся вершин дерева поиска надо выбрать такую, чтобы её значение было:

    больше всех значений в левом поддереве удаляемой вершины,
    меньше либо равно всем значениям в правом поддереве.

Где можно найти такую вершину? Для поиска правильного ответа мы рекомендуем нарисовать разные варианты деревьев поиска.
Тоже правильный ответ
Самую правую вершину в левом поддереве.
Там лежит самое большое значение из левого поддерева, и оно точно меньше, чем любое значение правого поддерева.
Самую левую вершину в левом поддереве.
Самую правую вершину в правом поддереве.
Правильный ответ
Самую левую вершину в правом поддереве.
Корень правого поддерева.
Корень левого поддерева.

После удаления вершины и замены её на другую подходящую дерево будет выглядеть одним из двух образов — в зависимости от того, будем мы брать вершину на замену из правого или из левого поддерева удалённого элемента
image
В данном случае на место удаляемой вершины мы поставили вершину из её правого поддерева со значением 13
Для простоты повествования мы всегда будем брать самую правую вершину в левом поддереве. При желании вы сможете адаптировать алгоритм, считая, что берётся самая левая вершина в правом поддереве.
Мы нашли вершину PPP, которая объединит распадающиеся поддеревья. Теперь надо переместить её на место удаляемой вершины DDD. Перемещение — это составное действие. Для краткости в этом алгоритме мы будем использовать следующие обозначения: P.parent — родитель вершины PPP, D.parent — родитель вершины DDD. Рассмотрим два варианта удаления:
Первый вариант
PPP — лист дерева, у него нет собственных детей.
image
Вершина P — лист. Значит её можно переместить на место вершины D, не опасаясь развала дерева
Это значит, что мы не оставим осиротевших поддеревьев, просто удалив её со старого места (P.parent.right = None). После удаления переместим вершину PPP на новое место. Алгоритм будет таким:

    Удаляем вершину со старого места. Для этого удалим ссылку на неё из P.parent. При этом мы знаем, что PPP — самая правая вершина, значит, и у своего родителя она будет правым ребёнком: P.parent.right = None.
    У родителя вершины DDD меняем ребёнка DDD на PPP. В этом месте мы не знаем, была ли вершина DDD правым или левым ребёнком своего родителя. Надо проверить каждого из детей D.parent, чтобы узнать, какого ребёнка заменит вершина PPP.
    Вершине PPP присваиваем детей DDD: левым ребёнком PPP станет левый ребёнок DDD (P.left = D.left), а правым — правый (P.right = D.right).

image
После удаления вершины D дерево поиска будет содержать на одну вершину меньше, но останется единым
Второй вариант
Если у вершины PPP есть потомки, значит, у неё есть только левый ребёнок и нет правого, так как PPP — самая правая вершина в своём поддереве.
image
Вершина P, которая должна заменить D, не является листом. Перенося её важно не развалить дерево на две части
Алгоритм действий:

    Чтобы не оставить P.left осиротевшим после переноса PPP, надо чтобы P.parent.right (прародитель P.left, родитель PPP) усыновил его вместо PPP.
    Теперь у вершины PPP нет детей, и её можно удалить, следуя алгоритму из первого варианта.

image
Вершина P.left стала правым ребёнком вершины P.parent. Никаких изменений с детьми вершины P.left проводить не требуется
Рита
Какой крутой алгоритм! Он работает, если дерево распалось и на два поддерева, и даже на три. Правда, удаляет он не все элементы с нужным ключом, а только первый попавшийся.

## Сбалансированные деревья поиска
[телепорт в оглавление](#title)

В этом спринте мы много говорили про деревья поиска. Рассмотрели, как происходят операции поиска, вставки и удаления. Напоминаем, что время выполнения операций в дереве поиска пропорционально его высоте HHH.
Если разработчик не умеет контролировать высоту дерева, она полностью зависит от того, в каком порядке добавляются вершины. Посмотрите на примеры неудачных деревьев поиска:
image
Высота каждого из этих деревьев равна количеству вершин в нём
Такие деревья поиска практически бесполезны, так как время выполнения основных операций в них O(H)=O(N)O(H)=O(N). С тем же успехом можно решать задачи с помощью связных списков.
Сбалансированность
Для того чтобы деревья давали ощутимый выигрыш по скорости, нужно минимизировать высоту дерева. В лучшем случае мы всегда будем добавлять новый элемент на минимально возможном удалении от корня.
Три вершины можно уместить на двух слоях бинарного дерева, семь — на трёх, а пятнадцать — на четырёх. Общая формула такая: если в дереве N вершин, то его высота будет не меньше ⌈log⁡2(N+1)⌉\lceil\log_2{(N+1)}\rceil⌈log2​(N+1)⌉, где специальные скобки ⌈...⌉\lceil ...\rceil⌈...⌉ обозначают округление вверх до ближайшего целого числа.
Представим, что мы ищем ключ в бинарном дереве поиска: рассматриваем каждый элемент, сравниваем его с искомым ключом и продолжаем поиск в левом либо в правом поддереве. Если после рассмотрения каждого из этих элементов оставшееся поддерево для поиска уменьшается примерно вдвое, то такое дерево поиска называется «сбалансированным» (англ. balanced binary search tree).
Сложность операции поиска по такому дереву равна O(log⁡2N)O(\log_2 N)O(log2​N), потому что на каждом шаге задача упрощается примерно в два раза. Соответственно, и высота сбалансированного бинарного дерева поиска H≈log⁡2NH \approx \log_2 NH≈log2​N.
В действительности необязательно требовать того, чтобы количество элементов уменьшалось именно в два раза (хотя это и является идеальной ситуацией). Вместо этого можно выбрать любое фиксированное число. Например, если на каждом шаге остаётся не больше, чем 9/109/109/10 элементов, то дерево будет иметь высоту log⁡10/9N\log_{10/9}{N}log10/9​N, что в несколько раз выше, чем log⁡2N\log_2{N}log2​N. Но эта высота всё равно будет логарифмически зависеть от числа элементов. Поскольку логарифмы с различным основанием отличаются в константное число раз, в асимптотической сложности алгоритма основание можно не указывать, таким образом H=O(log⁡N)H = O(\log{N})H=O(logN).
image
В сбалансированном дереве любую вершину можно найти не более, чем за 4 сравнения. В несбалансированном — худший случай потребует 7 сравнений
Критерии сбалансированности дерева
Существует несколько критериев, по которым можно проверить, сбалансировано дерево или нет. Увы, не все из них интуитивно понятны. Мы рассмотрим два основных критерия. Однако приводить доказательства, почему любой из них обеспечивает скорость поиска O(log⁡N)O(\log{N})O(logN), в рамках этого курса мы не будем.
При оценке сложности операций в деревьях поиска важно помнить про амортизированную сложность. Она состоит из двух частей:

    Сложность операций в сбалансированном дереве поиска. По определению она равна O(log⁡N)O(\log{N})O(logN).
    Сложность перебалансировки дерева, помноженная на частоту разбалансировки дерева.

Теперь давайте посмотрим на разные критерии сбалансированности. Как мы уже говорили, чтобы быть сбалансированным, дереву достаточно соответствовать хотя бы одному из них.
Критерий по количеству вершин
Чтобы скорость поиска в двоичном дереве поиска была O(log⁡N)O(\log{N})O(logN), каждый переход в правое или левое поддерево должен уменьшать количество вершин, среди которых мы ищем, примерно вдвое.
Это условие будет выполняться, если для каждого поддерева в бинарном дереве количество вершин в правом и левом детях будет равно с точностью +-1 вершина. Такое дерево будет называться «идеально сбалансированным двоичным деревом поиска».
Критерий 1. У любой вершины количество потомков в правом и левом детях различается не более, чем на единицу.
image
Поддерживать дерево в идеально сбалансированном состоянии слишком трудоёмко. Потому что при любой модифицирующей операции оно может утратить свою идеальную сбалансированность.
Посмотрите на пример. В этом дереве поиска есть 7 позиций, куда можно добавить новые данные. Но только одна из них подходит для вставки, которая не повредит идеальной сбалансированности дерева.
image
Ключ в выделенной вершине должен быть меньше ключа родителя (14), так как это левое поддерево родителя. Но не меньше ключа прародителя (12), так как для прародителя мы находимся в правом поддереве. Поэтому подходят только числа 12 и 13
Данные, попадающие на вход программе, редко способствуют поддержанию сбалансированности. Чем строже критерий сбалансированности, тем чаще дерево придётся перестраивать, чтобы оно продолжало ему соответствовать.
Критерий по высоте дерева
Критерий сбалансированности по количеству вершин, который мы разобрали выше, кажется интуитивно понятным. Но всё-таки большая часть сбалансированных деревьев поиска опирается на другой критерий — по высоте поддеревьев. Его понять сложнее.
Критерий 2. У любой вершины высота правого и левого поддеревьев различается не более, чем на единицу.
Дерево, которое удовлетворяет такому критерию, называют АВЛ-деревом.
image
Рассмотрим корень: высота левого поддерева 3, высота правого поддерева 4. Изучив все вершины, можно убедиться, что это сбалансированное дерево
Мы разобрали только два критерия, но есть и другие. Например, красно-чёрные деревья, Splay-деревья и проч.
Различные механизмы ребалансировки деревьев довольно сложны на практике, и рассказ о них выходит за рамки нашего курса. Большинству разработчиков не придётся писать код для ребалансировки даже при решении алгоритмических задач.

## Приоритетная очередь. Вставка и удаление
Для того чтобы поддерживать свойства бинарной кучи, надо понимать в каких ситуациях они могут нарушиться. В случае с пирамидой есть два модифицирующих действия, которые могут испортить структуру:

    вставка элемента в кучу,
    извлечение с удалением элемента из кучи.

В этом уроке мы разберём, как вставлять и удалять элементы в кучу, не нарушая её свойств.

Вставляя любой другой элемент, нужно проделывать операцию обмена до тех пор, пока куча не станет упорядоченной. В худшем случае новый элемент встанет на вершину пирамиды.

Теперь заберём самый приоритетный элемент из пирамиды. Для того чтобы в ней не появились дыры, на место извлечённого элемента поставим последний элемент из кучи:

Рассмотрим текущий узел и его дочерние элементы. Если текущий элемент — самый приоритетный из них, ничего не меняем.
```python
def sift_down(heap, idx):
    heap_len = len(heap) - 1
    left = idx * 2
    right = left + 1
#     print(heap, left, right)
    if heap_len < left:
        return idx

    if (right <= heap_len) and (heap[left] < heap[right]):
        max_idx = right
    else:
        max_idx = left


    if heap[idx] <  heap[max_idx]:
        heap[idx], heap[max_idx] = heap[max_idx], heap[idx]
        return sift_down(heap, max_idx)

    return idx
    
def test():
    sample = [-1, 12, 1, 8, 3, 4, 7]
    assert sift_down(sample, 2) == 5

# test()

def test2():
    arr = [-1, 90, 70, 50]
    arr2 = [-1, 9, 24, 49]
    print(sift_down(arr2, 1))
    print(arr2)
test2()
```

```python
# ID удачной попытки: 
# -- ПРИНЦИП РАБОТЫ --
# 1. Если элемент уже пустой, то значит удалять нечего, возвращаем none
# 
# 2. если не пустой, смотрим равно ли значение в данном узле искомому? Если да:
# Возвращаем правую ветку, если левая пустая. Правая тоже может быть в этот момент пустой, но ничего страшного)
# Если же левая не пустая, то смотрим пустая ли правая, если да, то без зазрения совести отдаем на выход левую.
# Всё-равно другой нет.
# 
# 3. Если же обе ветки есть, то идем до самого правого элемента левой ветки. Удаляем его из этой ветки, 
# рекурсивно вызвав эту же функцию для этой левой ветки и значения самого правого элемнта, который мы нашли
# и возвращаем эту ветку, где наверх добавили значение удаленного элемнта
# 
# 4. В случае если значение в узле оказалось не равно искомому, вызываем рекурсивно функцию для левой 
# или для правой ветки
# 

# -- ДОКАЗАТЕЛЬСТВО КОРРЕКТНОСТИ --
# если справа находятся элементы больше, чем родительский, а слева меньше, то самый правый элемент левой ветки,
# это самый большой элемент из тех, что меньше родительского, 
# а следовательно он не нарушит порядок остального дерева.
#     

# -- ВРЕМЕННАЯ СЛОЖНОСТЬ --
# Решение должно работать за O(H), где H глубина дерева, так как, мы сначала спускаемся на какую-то глубину h1,
# Когда ищем элемент, а потом будем искать крайний правый элемент в левой ветке и опустимся максимум на H - h1
# И для удаления мы пройдем второй раз H - h1. Итого O(H)

# -- ПРОСТРАНСТВЕННАЯ СЛОЖНОСТЬ --
# дополнительной памяти не нужно, глубина рекурсии в H только добавит потребления памяти. А так O(1)
# Не корректно выразился в первый раз: Глубина рекусрии потребует O(H) дополнительной памяти, 
# остальным в рамках аппроксимации мы пренебрегаем, так как это O() от константы.

def remove(root, key):
    if root is None:
        return None
    else:
        if root.value == key:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            else:
                next_item = root.left
                while next_item.right is not None:
                    next_item = next_item.right
                new = next_item.value
                root.value = new
                root.left = remove(root.left, new)
        elif key < root.value:
            root.left = remove(root.left, key)
        else:
            root.right = remove(root.right, key)

    return root
```
