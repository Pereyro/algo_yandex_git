## Построение хеш-функций для строк
[телепорт в оглавление](#title) 

Теперь давайте рассмотрим, как вычисляются хеш-функций для строк.
Для этого есть специальный алгоритм, который называется «полиномиальным хешированием» (от слова «полином» — многочлен). Хеш строки sss вычисляется по формуле:
h(s)=(s1qn−1+s2qn−2+⋯+sn−1q+sn) mod Rh(s)=\left(s_{1} q^{n-1}+s_{2} q^{n-2}+\cdots+s_{n-1} q+s_{n}\right) \bmod Rh(s)=(s1​qn−1+s2​qn−2+⋯+sn−1​q+sn​)modR
Где sis_isi​— код iii-го символа (например, ASCII-код), nnn — длина строки, а RRR и qqq — выбранные константы.
Все вычисления производятся по модулю некоторого большого числа RRR, поскольку для длинных строк сумма в скобках может оказаться слишком большой. В большинстве языков программирования при этом произойдёт переполнение типа целых чисел. В языках, где реализована так называемая «длинная арифметика», переполнения не случится, но с ростом значений складываемых и перемножаемых чисел вычисления будут становиться всё медленнее и медленнее.
Выбор числа RRR зависит от типа переменной, в которой хранятся хеши. Например, для беззнаковых 4-байтовых целых чисел удобно выбрать R=232.R=2^{32}.R=232. Для 8-байтовых чисел можно взять R=264.R=2^{64}.R=264.
Тимофей
В качестве qqq обычно берут какое-нибудь большое простое число.
Гоша
А что будет, если взять небольшое число?
Тимофей
Тогда хеши коротких строк будут находиться исключительно в области малых чисел. А нам, чтобы избежать коллизий, нужно распределить хеши как можно более равномерно.
Если строки длинные, выбор величины qqq не так важен — в сумму будут входить значения qqq в достаточно больших степенях. Но ради универсальности лучше и в этом случае взять большое число.
Например, часто используют число q=1000000007=109+7q = 1 000 000 007 = 10^9+ 7q=1000000007=109+7 , потому что оно большое, но всё ещё меньше, чем 2322^{32}232, и его легко запомнить.
Число qqq должно быть не только большим, но и простым.
Помните, мы говорили, что индекс корзины вычисляется взятием остатка от деления значения хеша на число корзин? То есть номер корзины в массиве длины MMM вычисляется как:
bucket(s)=((s1qn−1+s2qn−2+⋯+sn−1q+sn) mod R) mod M\mathrm{bucket}(s)=\left(\left(s_{1} q^{n-1}+s_{2} q^{n-2}+\cdots+s_{n-1} q+s_{n}\right) \bmod R\right) \bmod Mbucket(s)=((s1​qn−1+s2​qn−2+⋯+sn−1​q+sn​)modR)modM
Чем меньше общих делителей у qqq, MMM и RRR, тем реже будут возникать коллизии. В качестве RRR мы уже договорились использовать 2322^{32}232, поэтому любое нечётное число не будет иметь с RRR общих делителей.
Число MMM будет меняться в зависимости от размера хеш-таблицы. При вычислении хеш-функции этот размер ещё не известен. Из-за этого мы не можем гарантировать, что параметр вычисления хеш-функции qqq и число корзин MMM будут взаимно простыми. Но выбор большого простого числа в качестве qqq делает вероятность наличия нетривиальных общих делителей очень небольшой, ведь у простого числа нет делителей, кроме самого этого числа и единицы. Как мы говорили, MMM обычно тоже делают простым, поэтому общие делители могут получиться, только когда M=qM=qM=q.
Рассмотрим вычисление хеш-функции на примере строки "ABACB". ASCII-коды символов "A", "B" и "C" равны 65, 66 и 67.
В качестве констант возьмём: q=3q=3q=3, а R=28=256R=2^8=256R=28=256. Такие маленькие константы мы берём только для примера, чтобы число было проще посчитать.
Тогда значение хеш-функции для строки "ABACB" вычисляется так:
(65⋅34+66⋅33+65⋅32+67⋅31+66⋅30) mod 28=219.(65 \cdot 3^4 + 66 \cdot 3^3 + 65 \cdot 3^2 + 67 \cdot 3^1 + 66 \cdot 3^0) \bmod 2^8 = 219.(65⋅34+66⋅33+65⋅32+67⋅31+66⋅30)mod28=219.
Полиномиальный хеш можно эффективно вычислить с использованием метода Горнера:
h(s)=((((s1q+s2)q+s3)q+⋯+sn−1)q+sn) mod R.h(s)=\left(\left(\left(\left(s_{1} q+s_{2}\right) q+s_{3}\right) q+\cdots+s_{n-1}\right) q+s_{n}\right) \bmod R.h(s)=((((s1​q+s2​)q+s3​)q+⋯+sn−1​)q+sn​)modR.
В нашем случае метод Горнера можно применить так:
((((65⋅3+66)⋅3+65)⋅3+67)⋅3+66) mod 28=219.((((65 \cdot 3+66)\cdot 3 + 65) \cdot 3 + 67) \cdot 3+ 66) \bmod 2^8 = 219.((((65⋅3+66)⋅3+65)⋅3+67)⋅3+66)mod28=219.

Вычисление хешей методом скользящего окна
Давайте рассмотрим строку abcdefghij и её подстроки длины 6. Это будут abcdef, bcdefg, cdefgh, defghi, efghij.

abcdefghij
----------
abcdef
 bcdefg
  cdefgh
   defghi

Мы можем перебирать эти подстроки, сдвигаясь всякий раз на одну позицию вправо. Такой способ перебора называется методом «скользящего окна».
Обратите внимание, что подстроки перекрываются. Если взять две подстроки, начинающиеся с двух соседних позиций, мы увидим, что они перекрываются по K−2K-2K−2 позициям. В сдвинутой строке большая часть букв окажется такой же, как в исходной, и будет идти в том же порядке.
Запишем хеши для двух соседних строк:
h(‘‘abcdef")=(a~⋅q5+b~⋅q4+c~⋅q3+d~⋅q2+e~⋅q1+f~⋅q0)mod  Rh(``\mathrm{abcdef}") = \left( \tilde{\mathrm{a}}\cdot q^5 + \tilde{\mathrm{b}}\cdot q^4 + \tilde{\mathrm{c}}\cdot q^3 + \tilde{\mathrm{d}}\cdot q^2 + \tilde{\mathrm{e}}\cdot q^1 + \tilde{\mathrm{f}}\cdot q^0 \phantom{+\tilde{\mathrm{g}}\cdot q^0} \right) \mod Rh(‘‘abcdef")=(a~⋅q5+b~⋅q4+c~⋅q3+d~⋅q2+e~⋅q1+f~⋅q0+g~​⋅q0)modR
h(‘‘bcdefg")=(b~⋅q5+c~⋅q4+d~⋅q3+e~⋅q2+f~⋅q1+g~⋅q0)mod  Rh(``\mathrm{bcdefg}") = \left( \phantom{a\cdot q^5+} \tilde{\mathrm{b}}\cdot q^5 + \tilde{\mathrm{c}}\cdot q^4 + \tilde{\mathrm{d}}\cdot q^3 + \tilde{\mathrm{e}}\cdot q^2 + \tilde{\mathrm{f}}\cdot q^1 + \tilde{\mathrm{g}}\cdot q^0\right) \mod Rh(‘‘bcdefg")=(a⋅q5+b~⋅q5+c~⋅q4+d~⋅q3+e~⋅q2+f~⋅q1+g~​⋅q0)modR
Второй хеш можно выразить через первый с небольшими изменениями:
h(‘‘bcdefg")=((h(‘‘abcdef")−a~⋅q5)⋅q+g~)mod  Rh(``\mathrm{bcdefg}") = \left(\left(h(``\mathrm{abcdef}") - \tilde{\mathrm{a}}\cdot q^5\right) \cdot q + \tilde{\mathrm{g}}\right) \mod Rh(‘‘bcdefg")=((h(‘‘abcdef")−a~⋅q5)⋅q+g~​)modR
Таким образом, пересчёт хеша при сдвиге окна на одну позицию требует всего лишь O(1)O(1)O(1) операций. А значит, подсчёт хешей всех подстрок можно провести за O(L)O(L)O(L).

```python
# Схема Горнера
def polihash(string, q, m):
    hash_sum = 0
    for letter in string:
        hash_sum *= q
        hash_sum += ord(letter)
        hash_sum %= m
    

    return hash_sum

a = int(input())
m = int(input())
string = input()

print(polihash(string, a, m))
```

Хеш-таблица должна поддерживать следующие операции:

    Разрешать коллизии следует с помощью метода цепочек или с помощью открытой адресации.
    Все операции должны выполняться за O(1) в среднем.
    Поддерживать рехеширование и масштабирование хеш-таблицы не требуется.
    Ключи и значения, id сотрудников и их зарплата, —– целые числа. Поддерживать произвольные хешируемые типы не требуется.

```python
# ID удачной попытки: 
# -- ПРИНЦИП РАБОТЫ --
# 1. Будем использовать в качестве хеша от значения, само значение, так как на вход подаются натуральные числа.
# Поиск индекса будем осуществлять через остаток от деления на количество ячеек.
# В качестве размера исходного массива (количество ячеек) мы возьмем большое простое число.
# Простое число было выбрано наугад
# Коллизии разрешаются методом связного списка
# 

# -- ДОКАЗАТЕЛЬСТВО КОРРЕКТНОСТИ --
# 
#     

# -- ВРЕМЕННАЯ СЛОЖНОСТЬ --
# Добавление нового элемента происходит за O(1). Так как мы выбрали большое простое число
# для размера исходного массива, то распределение должно быть более менее равномерным. 
# Следовательно для 10^5 ключей, длина связного списка должна быть в среднем 5 элементов
# 

# -- ПРОСТРАНСТВЕННАЯ СЛОЖНОСТЬ --
# По памяти при количестве ячеек M словарь должен занимать порядка O(M), но если количество ключей n > M,
# то по памяти словарь будет занимать O(n)
# 


if __name__ == '__main__':
    
    class Node:  
        def __init__(self, value, next_item=None):  
            self.value = value  
            self.next_item = next_item
    
    class Hashtable:
        def __init__(self, map_size):
            self.map_size = map_size
            self.array = [None] * map_size

            
        def put(self, key, new_value):
            index = key % self.map_size
            node0 = self.array[index]
#             print('index:', index)
            node = node0
            stop = True
            while node is not None:
#                 print('nodevalue', node.value)
                if node.value[0] == key:
                    node.value = [key, new_value]
                    break
#                     print('new value', node.value)
                else:
                    node = node.next_item
            
            if node is None:
                self.array[index] = Node([key, new_value], node0)
#             print('nodevalue fin', self.array[index].value)
            
            
        def get(self, key):
            index = key % self.map_size
            node = self.array[index]
            while node is not None:
#                 print('nodevalue', node.value)
                if node.value[0] == key:
                    return node.value[1]
                else:
                    node = node.next_item
            return None
                                
        def delete(self, key):
            index = key % self.map_size
            node = self.array[index]
            if node is None:  
                return None
            else:
                if node.value[0] == key:
                    out_value = node.value[1]
                    self.array[index] = node.next_item
                    return out_value
                else:
                    while node.next_item is not None:
#                         print('nodevalue', node.value)
                        if node.next_item.value[0] == key:
                            out_value = node.next_item.value[1]
                            node.next_item = node.next_item.next_item
                            return out_value
                        else:
                            node = node.next_item
            return None    
        
                
    testmode = False
    if testmode:
        input_list = ['put 1 2', 'put 7 8', 'put 8 9', 'put 101 102', 'put 103 104', 'put 23 21', 'get 5', 'delete 7']
        comand_count = len(input_list)
        maparray = Hashtable(7)
        
        out_array = []
        for comand_str in input_list:
            comand, *values = comand_str.split()
#             print(comand, values)
            if comand == 'delete':
                out_array.append(maparray.delete(int(*values)))          
            elif comand == 'get':
                out_array.append(maparray.get(int(*values)))
            else:
                maparray.put(*map(int,values))
                
        print(*out_array, sep='\n')
    else:
        comand_count = int(input())
        maparray = Hashtable(23011)
        
        out_array = []
        for i in range(comand_count):
            comand, *values = input().split()
#             print(comand, values)
            if comand == 'delete':
                out_array.append(maparray.delete(int(*values)))          
            elif comand == 'get':
                out_array.append(maparray.get(int(*values)))
            else:
                maparray.put(*map(int,values))
                
        print(*out_array, sep='\n')
```
